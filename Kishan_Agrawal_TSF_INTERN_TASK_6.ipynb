**What is Iris Dataset?**

*   The Iris Dataset contains four features (length and width of sepals and petals) of 50 samples of three species of Iris (Iris Setosa, Iris Virginica and Iris Versicolor). These measures were used to create a linear discriminant model to classify the species. The dataset is often used in data mining, classification and clustering examples and to test algorithms.

**Importing Libraries**

```python
#importing libraries
import pandas as pd
import numpy as np
import seaborn as sns
sns.set_palette('husl')
import graphviz

import matplotlib.pyplot as plt
%matplotlib inline
import pydot
from IPython.display import Image

from sklearn.tree import DecisionTreeClassifier,export_graphviz,plot_tree
from sklearn.model_selection import train_test_split
import sklearn.metrics as sm
```

**Reading the Data.**

```python
iris_data = pd.read_csv("https://raw.githubusercontent.com/mrkishanagrawal/GRIP-OCT-22-TASK-06/main/Iris.csv")
```

```python
iris_data.head()
```

**Dropping 'Id' Column as it is not necessary**

```python
iris_data.drop('Id',axis=1,inplace=True)
```

**Pair-Plotting different combinations from data**

```python
plt.close()
sns.set_style('whitegrid')
sns.pairplot(iris_data,hue='Species',height=3)
plt.show()
```

**Plotting SepalLength v/s SepalWidth and PetalLength v/s PetalWidth**

```python
sns.FacetGrid(iris_data,hue='Species').map(plt.scatter,'SepalLengthCm','SepalWidthCm').add_legend()
plt.show()
sns.FacetGrid(iris_data,hue='Species').map(plt.scatter,'PetalLengthCm','PetalWidthCm').add_legend()
plt.show()
```

**Storing data and target in X and y respectively**

```python
X = iris_data.iloc[:,:-1]
y = iris_data['Species']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=101)
tree_obj = DecisionTreeClassifier()
tree_obj.fit(X_train,y_train)
print("Training Complete !")
y_pred = tree_obj.predict(X_test)
```

**Comparing Actual and Predicted Value**

```python
df = pd.DataFrame({'Actual':y_test,'Predicted':y_pred})
df
```

**Decision Tree Graph**

```python
features=['Sepal Length','Sepal Width','Petal Length','Petal Width']
class_=['Setosa', 'Versicolor', 'Virginica']

fig,axes = plt.subplots(nrows=1,ncols=1,figsize=(4,4),dpi=350)
plot_tree(tree_obj,feature_names=features,class_names=class_,filled=True,rounded=True)
```

**Prediction with random values**

```python
print('Class Names : ',class_,"\n")
print('Enter the following values :')
sepal_length=float(input('Enter Sepal Length : '))
sepal_width =float(input('Enter Sepal Width  : '))
petal_length=float(input('Enter Petal Length : '))
petal_width =float(input('Enter Petal Width  : '))
print(f'\nEstimating the Class Probalicties.....')
output = tree_obj.predict([[sepal_length, sepal_width,petal_length,petal_width]])
# output = tree_obj.predict([[5.1, 3.4,1.4,0.2]])
print('Output : ',output[0])
```

**Accuracy of Model**

```python
# Model Accuracy
print('Accuracy : ',sm.accuracy_score(y_test,y_pred))
```
